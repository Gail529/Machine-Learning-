```
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data=pd.read_excel('/content/Australia2.xlsx' ,sheet_name='Sheet1')
```

```
data
```

```
plt.style.use('seaborn')
date=data['Unnamed: 0']
act=data['ACT']
nsw=data['NSW']
nt=data['NT']
qld=data['QLD']
plt.plot_date(date,act ,  linestyle='solid' ,color='blue', label='Australian Capital Territory')  
plt.plot_date(date,nsw  ,linestyle='solid', color='grey' , label='New South Wales')  
plt.plot_date(date,nt,linestyle='solid' ,color='pink', label='Nothern Territory')  
plt.plot_date(date,qld,linestyle='solid' ,color='purple', label='Queensland')  
plt.legend()
plt.tight_layout()
plt.show()
```

```
plt.style.use('seaborn')
date=data['Unnamed: 0']
sa=data['SA']
tas=data['TAS']
vic=data['VIC']
wa=data['WA']
plt.plot_date(date,sa ,  linestyle='solid' ,color='blue', label='South Australia')  
plt.plot_date(date,tas  ,linestyle='solid', color='grey' , label='Tasmania')  
plt.plot_date(date,vic,linestyle='solid' ,color='pink', label='Victoria')
plt.plot_date(date,wa,linestyle='solid' ,color='purple', label='Western Australia') 
plt.legend()
plt.tight_layout()
plt.show()
```

```
data2=pd.read_excel('/content/Australia2.xlsx',sheet_name='Sheet3',skiprows=2)
```

```
total=data2['total E']
male=data2['Male E']
female=data2['Female E']
utotal=data2['total U']
umale=data2['Male U']
ufemale=data2['Female U']
urtotal=data2['total rate']
urmale=data2['Male rate']
urfemale=data2['Female rate']
```

```
labels=['2019-10','2019-11','2019-12','2020-01','2020-02','2020-03','2020-04']
x=np.arange(len(labels))
width = 0.2
fig,   ax=plt.subplots()
plt.style.use('seaborn')
ax.bar(x-width,   total,width,label='Total' ,color='brown',align='center')
ax.bar(x,  male,width,label='Men' ,color='grey',align='center')
ax.bar(x+width,  female,width,label='Women' ,color='yellow',align='center')
ax.set_ylabel('No of employed Persons')
ax.set_title('Employed  persons Austria ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()
plt.show()
```

```
labels=['2019-10','2019-11','2019-12','2020-01','2020-02','2020-03','2020-04']
x=np.arange(len(labels))
width = 0.35
fig,   ax=plt.subplots()
plt.style.use('seaborn')
ax.bar(x-width/2,   utotal,width,label='Total' ,color='blue')
ax.bar(x+width/2,  umale,width,label='Men' ,color='grey')
ax.bar(x+width/3,  ufemale,width,label='Women' ,color='black')
ax.set_ylabel('No of  Unemployed Persons')
ax.set_title('Unemployed  persons Austria ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()
plt.show()
```

```
import datetime as dt
import matplotlib.dates as mdates
```

```
string=policy['StringencyIndex']
date=policy['Date']
plt.style.use('seaborn')
plt.figure(figsize=(12,5))
plt.plot(date,string,color='grey',linestyle='dashdot')
plt.annotate('Canceling of  all public events  \n Restrictions on gatherings(limit 100-1000 people)  \n (17/03/2020)' ,(mdates.date2num(date[17]) ,(string[17] )  )  ) 
plt.annotate(' (24/03/2020) \n Workplace closing \n   Restrictions on gatherings(limit 10-100 people)' ,(mdates.date2num(date[23]) ,(string[23] )  )  ) 
plt.annotate(' Restrictions on gatherings(limit 10 people) \n (29/03/2020) ' ,(mdates.date2num(date[30]) ,(string[30] )  )  ) 
plt.annotate(' Relaxed workplace policy  \n  to specific sectors\n(18/04/2020)' ,(mdates.date2num(date[53]) ,(string[53] )  )  ) 
plt.xlabel('Date')
plt.ylabel('StringencyIndex')
plt.tight_layout()
plt.legend()
plt.show()

```
