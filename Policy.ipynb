```
plt.style.use('seaborn')
date=data['Unnamed: 0']
construction=data['Construction']
manufacture=data['Manufacture ']
trade=data['Trade']
transport=data['Transportation and storage']
accomod=data['Accomodation and gastronomy']
health=data['Health and Social care']
plt.plot_date(date,   manufacture,   linestyle='solid' ,color='black', label='manufacture')
plt.plot_date(date,   construction,   linestyle='solid' ,color='pink', label='Construction')
plt.plot_date(date,   trade,   linestyle='solid' ,color='blue', label='Trade')
plt.plot_date(date,   transport,   linestyle='solid' ,color='brown', label='Transport')
plt.plot_date(date,   accomod,   linestyle='solid' ,color='yellow', label='Accomodation')
plt.plot_date(date,   manufacture,   linestyle='solid' ,color='grey', label='manufacture')
plt.plot_date(date,   health,   linestyle='solid' ,color='purple', label='Health and Social care')
plt.annotate('lowest value' ,(mdates.date2num(date[3]) ,(accomod[3] )  ),  (mdates.date2num(date[4]) ,(accomod[4] )  ) ,
  arrowprops=dict(facecolor='black', shrink=0.3 ))
plt.title('Austria')
plt.xlabel('date')
plt.ylabel('unemployed persons')
plt.legend()
plt.show()
```
